```python
'''
    TÍNH ĐÓNG GÓI
    ta đang để mặc định các phương thức và lớp là public, nên có thể truy cập được cả từ bên ngoài
    và lớp con
    
'''
class Animal1:
    
    # khai báo name và age là public
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # phương thức trừu tượng
    def make_sound(self):
        pass 
    
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

animal1 = Animal1("Dog", 3)
animal1.display_info()
print(animal1.name)
print(animal1.age)

print("------------------------------------")

# cú pháp tạo lớp con: Tên_lớp_con(Tên_lớp_cha)
class Dog1(Animal1):
    def make_sound(self):
        print("Bark bark")

dog1 = Dog1("Buddy", 2)
dog1.display_info()    
dog1.make_sound()     

    
    
```

```python
'''
    TÍNH ĐÓNG GÓI
    protected: được quy được bằng_
    Tuy nhiên theo ví dụ, ta vẫn có thể truy cập trực tiếp vào biến dù nó đã protected
    Nói chung single underscore chỉ mang ý nghĩa cảnh báo chứ không có tính năng
    
'''
class Animal2:
    
    # khai báo name và age là public
    def __init__(self, name, age):
        self._name = name
        self._age = age
    
    # phương thức trừu tượng
    def make_sound(self):
        pass 
    
    # public method
    def display_info(self):
        print(f"Name: {self._name}")
        print(f"Age: {self._age}")

animal1 = Animal2("Dog", 3)
animal1.display_info()
print(animal1._name)
print(animal1._age)

print("------------------------------------")

class Dog2(Animal2):
    def make_sound(self):
        print("Bark bark")

dog1 = Dog2("Buddy", 2)
dog1.display_info()    
dog1.make_sound()     

    
    
```

```python
'''
    TÍNH ĐÓNG GÓI
    private: được quy được bằng __
    Với private, ta không thể truy cập trực tiếp được dữ liệu,
    mà bắt buộc vẫn phải thông quá public method display_info()
    Cố gắng truy cập thẳng vào dữ liệu private sẽ bị lỗi
    Thường người ta sẽ tạo ra các public getter à setter
    
    
'''
class Animal3:
    
    # khai báo name và age là public
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    
    # phương thức trừu tượng
    def __make_sound(self):
        print("go")
    
    # public method
    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")

animal1 = Animal3("Dog", 3)
animal1.display_info()
print(animal1.__name)
print(animal1.__age)
```

```python
'''
    TÍNH KẾ THỪA
    super() được sử dụng để kế thừa lại các thuộc tính từ lớp cha 
'''
```

```python

class Vehicle:
    def __init__(self, name, age, price):
        self.__name = name
        self.__age = age
        self.__price = price
    
    def get_age(self):
        return self.__age
    
    def get_name(self):
        return self.__name
    
    def get_price(self):
        return self.__price
    
    
# Xây dựng class con dựa trên class cha
# Bổ sung thêm thuộc tính wheel
class Motorbike(Vehicle):
    def __init__(self, name, age, price):
        super().__init__(name, age, price)
        self.__wheels = 4
        
    def get_n_wheels(self):
        return self.__wheels


        

```

```python
'''
    như vậy lớp con Motorbike hoàn toàn kế thừa các phương thức từ lớp cha
    
'''
vehicle = Vehicle("Xe Chung", 15, 15)
print(vehicle.get_name())
print(vehicle.get_age())
print(vehicle.get_price())

print("--------------------------------")

motorbike  = Motorbike("Xe Máy", 2, 10)
print(motorbike.get_name())
print(motorbike.get_age())
print(motorbike.get_price())
print(motorbike.get_n_wheels())
```

```python
class Guitar:
    __n_strings = 6
    __height = 10
    def __init__(self):
        pass
    
    def get_strings(self):
        return self.__n_strings
    
    def get_height(self):
        return self.__height

class ElectricGuitar(Guitar):
    def __init__(self):
        super().__init__()
        self.__brand = "Yamaha"
        self.__height = 8
    
    # nếu không thêm hàm này, giá trị của height ở e_guitar sẽ vẫn là 10
    # điều này xảy ra vì python không tìm thấy height nào trong lớp con (do chưa khởi tạo trong init)
    
    # đây là ví dụ nhỏ của ghi đè
    def get_height(self):
        return self.__height
    
    def get_brand(self):
        return self.__brand
        
        
```

```python
guitar = Guitar()
print(guitar.get_strings())
print(guitar.get_height())

print("--------------------------")

e_guitar = ElectricGuitar()
print(e_guitar.get_strings())
print(e_guitar.get_height())
print(e_guitar.get_brand())

```

```python
'''
    Thử sử dụng overload trong Python
'''
class Animal:
    def __init__(self, name, height, weight):
        self.__name = name
        self.__height = height
        self.__weight = weight
    
    def get_name(self):
        return self.__name
    
    # 2 hàm này, hàm xuất hiện sau cùng sẽ được chọn
    # Python không hỗ trợ overload như trong Java
    def make_sound(self, a):
        print("Animals are making sound", a)
        
    def make_sound(self):
        print("Animals are making sound")
    


```

```python
# bị lỗi vì python không hỗ trợ overload
animal = Animal("Dog", 14, 14)
animal.make_sound(5)
```

```python
'''
    Overriding
'''
class Animal:
    def __init__(self, name, height, weight):
        self.__name = name
        self.__height = height
        self.__weight = weight
    
    def get_name(self):
        return self.__name
        
    def introduce(self):
        print("I'm " +self.__name+ " . I'm an animal")
```

```python
animal = Animal("Ben", 20, 20)
animal.introduce()
```

```python
# tạo 2 lớp cat và dog thừa kế từ animal
# mỗi loài này có một phương thức make_sound khác nhau
class Dog(Animal):
    def __init__(self, name, height, weight):
        super().__init__(name, height, weight)
    
    # Override Method
    def introduce(self):
        print("I'm " +super().get_name()+ " . I'm a Dog")
    

class Cat(Animal):
    def __init__(self, name, height, weight):
        super().__init__(name, height, weight)
    
    # Override Method
    # Dùng super() vì phương thức đó ở lớp cha
    def introduce(self):
        print("I'm " +super().get_name()+ " . I'm a Cat")
```

```python
dog = Dog("Kun", 20, 20)
dog.introduce()

cat = Cat("Catmeows", 20, 20)
cat.introduce()
```

```python
'''
    TÍNH TRỪU TƯỢNG
    Lớp trừu tượng chỉ có tên phương thức
    Các lớp con kế thừa cần ghi đè hết các phương thức trừu tượng
    Trong python, cần phải thực hiện thông qua abtract class abc
    @ được gọi là decorator
    
'''
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
    
    def introduce(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return "Bark"

class Cat(Animal):
    def make_sound(self):
        return "Meow"

def animal_sound(animal: Animal):
    print(animal.make_sound())

dog = Dog()
cat = Cat()

animal_sound(dog) 
animal_sound(cat) 
```
